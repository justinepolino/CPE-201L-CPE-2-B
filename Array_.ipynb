```python
def traverse_array(arr):
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
    reversed_arr = arr[::-1]
    print(reversed_arr)

def sort(arr):
    arr.sort()
    print("Sorted array:", arr)

def insert(arr):
    value = int(input("Enter value to insert: "))
    index = input("Enter index to insert at: ")
    if index:
        arr.insert(int(index), value)
    else:
        arr.append(value)
    print("Array after insertion:", arr)

def delete(arr):
    index = int(input("Enter index to delete: "))
    arr.pop(index)
    print("Array after deletion:", arr)

def modify(arr):
    index = int(input("Enter index to modify: "))
    value = int(input("Enter new value: "))
    arr[index] = value
    print("Array after modification:", arr)

def main():
    arr = [10, 60, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Sort array")
        print("4. Insert element")
        print("5. Delete element")
        print("6. Modify element")
        print("7. Exit")
        choice = input("Enter your choice (1 to 7): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse(arr)
        elif choice == '3':
            sort(arr)
        elif choice == '4':
            insert(arr)
        elif choice == '5':
            delete(arr)
        elif choice == '6':
            modify(arr)
        elif choice == '7':
            break
        else:
            print("Invalid choice. Please enter a number between 1 to 7.")

if __name__ == "__main__":
    main()
```
